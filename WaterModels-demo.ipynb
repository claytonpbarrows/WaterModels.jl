````julia
using WaterModels
using InfrastructureModels
using JuMP
using Ipopt
using GLPKMathProgInterface
using JSON
````

````julia
glpk = GLPKSolverMIP(presolve = false, msg_lev = GLPK.MSG_ON)
ipopt = IpoptSolver(print_level = 1, tol = 1.0e-9)
````

````julia
datapath = joinpath(dirname(dirname(pathof(WaterModels))),"test/data/epanet")
jsonpath = joinpath(dirname(dirname(pathof(WaterModels))),"test/data/json")
network = "balerma"
````

````julia
network_path = joinpath(datapath,join([network,".inp"]))
modification_path = joinpath(jsonpath,join([network,".json"]))
````

I'm not sure why, but if the .json file doesn't exist, you need to parse the .inp file and then create the .json to run any WaterModels.jl simulations

````julia
wd = WaterModels.parse_file(network_path)
````

````julia
open(modification_path, "w") do f
    JSON.print(f, wd)
end
````

````julia
mmilp, rnlp = WaterModels.initialize_models(network_path, modification_path)
````

````julia
mmilp.model
````

````julia
rnlp.model
````

````julia
status = solve_global(network_path, modification_path, ipopt, glpk)
````
